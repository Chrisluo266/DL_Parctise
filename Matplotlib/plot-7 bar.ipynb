```python
import pylab as plt
import numpy as np

point_num = 30
x = np.arange(point_num)
y1 = np.random.uniform(0,50,point_num)
y2 = np.random.uniform(0,50,point_num)

plt.figure(figsize=(20,20))
bar_res = plt.bar(x,y1,color='#9999ff',edgecolor='k')
plt.bar(x,-y2,color='#ff9999',edgecolor='k')
print(bar_res)

for xv,y1v,y2v in zip(x,y1,y2):
    plt.text(xv,y1v + 2,"%d"%y1v,fontsize=16,ha="center",va="top")
    plt.text(xv,-(y2v + 2),"%d"%-y2v,fontsize=16,ha="center",va="bottom")
plt.show()
```
