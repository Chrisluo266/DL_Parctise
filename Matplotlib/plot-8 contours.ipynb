```python
import pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def height(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 100
X = np.linspace(-5,5,n)
Y = np.linspace(-5,5,n)
X,Y = np.meshgrid(X,Y)
Z = height(X,Y)

print("X.shape = ",X.shape)
print("Y.shape = ",Y.shape)
print("Z.shape = ",Z.shape)

plt.figure(figsize=(10,6.18))
plt.contourf(X,Y,Z,8,alpha=0.75,cmap=plt.cm.Greens)
line = plt.contour(X,Y,Z,8,colors="k",linewidths=0.5)
plt.clabel(line,inline=True,fontsize=10)
plt.axis("off")
plt.show()

fig = plt.figure(figsize=(10,6.18))
axes3d = Axes3D(fig)
surf = axes3d.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap="rainbow",edgecolor="black")
#axes3d.contourf(X,Y,Z,zdir="z",offset=-1.5,cmap='rainbow')
plt.show()
```
