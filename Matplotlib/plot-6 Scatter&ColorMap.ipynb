```python
import pylab as plt
import numpy as np
import matplotlib
import matplotlib.cm as cm 
```

```python
point_num = 1024
x = np.random.uniform(-50,50,point_num)
y = np.random.uniform(-50,50,point_num)
color = np.arctan2(x,y)
print(color.shape)
#colorlistColorMap(viridis)0-1rgba
args = plt.scatter(x,y,c=color,alpha=0.5,cmap='rainbow')
print(args)
print(args.cmap.name)

#colormap0-1rgbalist
plt.figure()
minima = color.min()
maxima = color.max()
norm = matplotlib.colors.Normalize(vmin=minima, vmax=maxima, clip=False) 
mapper = cm.ScalarMappable(norm=norm, cmap=args.cmap) 

color2 = mapper.to_rgba(color)
print("color.shape = ",color.shape)
print("color2.shape = ",color2.shape)
plt.scatter(x,y,color=color2,alpha=0.5)
#plt.colorbar()
plt.show()
```
