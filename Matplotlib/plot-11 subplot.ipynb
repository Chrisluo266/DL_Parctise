```python
#%matplotlib notebook
import pylab as plt
import numpy as np

plt.figure()
#分成3*2,第三位是编号  也可以写成321的形式
plt.subplot(321)
plt.plot([0,1],[0,1])

plt.subplot(3,2,2)
plt.plot([0,1],[0,2])

plt.subplot(3,2,3)
plt.plot([0,1],[0,3])

plt.subplot(3,2,4)
plt.plot([0,1],[0,4])

plt.subplot(3,2,5)
plt.plot([0,1],[0,5])

plt.subplot(3,2,6)
plt.plot([0,1],[0,6])
plt.show()
```

```python
plt.figure()
#大的尺寸是2*3，
#第一个表占2*1，编号为1，实际占用编号1,2,3
plt.subplot(2,1,1)
plt.plot([0,1],[0,1])

#第二个表占2*3，编号为4，因为表一占了1,2,3三格，所以从4开始
plt.subplot(2,3,4)
plt.plot([0,1],[0,2])

plt.subplot(2,3,5)
plt.plot([0,1],[0,3])

plt.subplot(2,3,6)
plt.plot([0,1],[0,4])

plt.show()
```
