```python
import pylab as plt
import numpy as np
```

```python
#method 1:subplot to grid
plt.figure()
#shape,index
ax1 = plt.subplot2grid((3,3),(0,0),colspan=3)
ax1.plot((1,2),(1,2))
ax1.set_title("Ax1")
ax2 = plt.subplot2grid((3,3),(1,0),rowspan=2)
ax2.plot((1,2),(1,2))
ax2.set_title("Ax2")
ax3 = plt.subplot2grid((3,3),(1,1),colspan=2)
ax3.plot((1,2),(1,2))
ax3.set_title("Ax3")
ax4 = plt.subplot2grid((3,3),(2,1))
ax4.plot((1,2),(1,2))
ax4.set_title("Ax4")
ax5 = plt.subplot2grid((3,3),(2,2))
ax5.plot((1,2),(1,2))
ax5.set_title("Ax5")

plt.show()
```

```python
import matplotlib.gridspec as gridspec
#method 2:sgridspec
plt.figure()
# 3*3
gs = gridspec.GridSpec(3,3)
print(gs[0,:])
ax1 = plt.subplot(gs[0,:])
ax1.plot((1,2),(1,2))
ax1.set_title("Ax1")
ax2 = plt.subplot(gs[1:,0])
ax2.plot((1,2),(1,2))
ax2.set_title("Ax2")
ax3 = plt.subplot(gs[1,1:])
ax3.plot((1,2),(1,2))
ax3.set_title("Ax3")
ax4 = plt.subplot(gs[2,1])
ax4.plot((1,2),(1,2))
ax4.set_title("Ax4")
ax5 = plt.subplot(gs[2,2])
ax5.plot((1,2),(1,2))
ax5.set_title("Ax5")

plt.show()
```
