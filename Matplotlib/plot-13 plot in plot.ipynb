```python
import pylab as plt
import numpy as np
```

```python
fig = plt.figure()
x = np.arange(1,8)
y = [7,5,1,2,5,7,1]

plt.plot((0,10),(0,10))

l,b,w,h = 0.2,0.2,0.5,0.5
ax1 = fig.add_axes([l,b,w,h])
ax1.plot(x,y,color='r')
ax1.set_title("ax1")

l,b,w,h = 0.35,0.50,0.125,0.125
ax2 = fig.add_axes([l,b,w,h])
ax2.plot(y,x,color='b')
ax2.set_title("ax2")

plt.axes([0.5,0.25,0.125,0.125])
plt.plot(y[::-1],x,'g')
ax2.set_title("ax3")

plt.show()
```
