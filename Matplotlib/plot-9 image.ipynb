```python
import pylab as plt
import numpy as np
```

```python
a = np.array([0.3,0.36,0.4,0.36,0.4,0.5,0.4,0.5,0.66]).reshape(3,3)

print(a.shape)
plt.axis("off")
plt.imshow(a,interpolation='nearest',cmap=plt.cm.Greens)
plt.colorbar(shrink=0.9)

print("生成一张噪点图")
plt.figure()
image = np.random.normal(0.5,0.5,(32,32,3))
image = image.clip(0,1)
plt.axis("off")
plt.imshow(image)
plt.show()
```
