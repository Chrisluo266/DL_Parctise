```python
import pylab as plt
import numpy as np
x1 = np.linspace(-1,1,100)
y1 = 10*x1**2-5

x2 = x1
y2 = 2*x1+1

plt.figure(figsize=[10,5])
plt.plot(x1,y1,'.-',linewidth=1,label="y=10x^2-5")
plt.plot(x2,y2,color='#ff0000',linestyle='--',label = 'y=2x+1')
#plt.plot中使用label定义，曲线的label,然后再使用plt.legend()
plt.legend(loc='upper center')
#设置轴label
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
y_ticks = np.arange(-5,6,1)
plt.yticks(y_ticks)

#####
plt.figure()
plot1 = plt.plot(x1,y1,linewidth=0.5,label="y=10x^2-5")   #由于返回是一个Line2D列表，一般只包含一个元素
plot2 = plt.plot(x2,y2,color='#ff0000',linestyle='--',label = 'y=2x+1')
print(plot1,"    len = ",len(plot1))
print(plot2,"    len = ",len(plot2))
l1 = plot1[0]
l2 = plot2[0]
print(l1)
print(l2)
#这里label的优先级高于上面plot中label的优先级，图中的顺序（从上到下）按照handles的顺序，labels和handles对应
plt.legend(handles=[l2,l1],labels=["line","conic"],loc=3)   

plt.show()
```
