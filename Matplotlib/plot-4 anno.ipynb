```python
import pylab as plt
import numpy as np
```

```python
x1 = np.linspace(-3,3,121)
y1 = x1**2-4

x2 = x1
y2 = 2*x2-5

#挪动轴的位置
plt.figure()
plt.plot(x1,y1)
plt.plot(x2,y2,color='#ff0000')

#get_current_axis
ax = plt.gca()
spines = ax.spines
spines["right"].set_color("None")
spines["top"].set_color("None")

spines["bottom"].set_position(("data",0))
spines["left"].set_position(("data",0))

#相交点坐标是(1,-3)
plt.scatter(1,-3,s=50,color='#00ff00')
#画绿点的虚线
plt.plot([1,1],[0,-3],linestyle='--')        #(1,0)--->(1,-3)
plt.plot([0,1],[-3,-3],linestyle='--')       #(0,-3)--->(1,-3)

#写注释，method 1
plt.annotate("(%d,%d)"%(1,-3),fontsize=16,xy=(1,-3),xycoords="data",xytext=(40,-40),textcoords='offset points',
            arrowprops={
                'arrowstyle':'->',
                'connectionstyle':'arc3,rad=0.5',
            })
#写注释，method 2
plt.text(-2,1,"$\mu,\\alpha_t,\\sigma_i$",fontsize=16,color='#ff00ff')

plt.show()
```
