```python
import pylab as plt
import numpy as np

x = np.arange(1,20)
y1 = 3*x**2-5
y2 = np.sin(x)

fig,ax1 = plt.subplots()   #类似fig = plt.figure()   ax1 = plt.gca()
#fig = plt.figure()
#ax1 = plt.gca()
print("gca is ax1:",plt.gca() == ax1)

ax2 = ax1.twinx()   #将ax1的Y镜面出ax2的Y ,保持X不变
ax1.plot(x,y1,"g-")
ax2.plot(x,y2,"b-")

ax1.set_ylabel("Y1",color="g",fontsize=16)
ax2.set_ylabel("Y2",color="b",fontsize=16)

print("gca is ax1:",plt.gca() == ax1)
print("gca is ax2:",plt.gca() == ax2)
plt.show()
```
