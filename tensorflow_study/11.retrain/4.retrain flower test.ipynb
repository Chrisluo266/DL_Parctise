```python
import tensorflow as tf
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
lines = tf.gfile.GFile("./graph/output_labels.txt","r").readlines()

uid_to_flowers_map = {}

for uid,line in enumerate(lines):
    line = line.strip("\n")
    uid_to_flowers_map[uid] = line

def id_to_classname(uid):
    global uid_to_flowers_map
    if uid not in uid_to_flowers_map:
        return "Unknown"
    else:
        return uid_to_flowers_map[uid]

#读取pb
with tf.gfile.GFile("./graph/output_graph.pb","rb") as pbf:
    graph_def = tf.GraphDef()
    graph_def.ParseFromString(pbf.read())
    tf.import_graph_def(graph_def)
```

```python
with tf.Session() as sess:
    with tf.summary.FileWriter("logs/",sess.graph) as writer:
        print("logs")
```

<p>读取test tfrecords文件</p>

```python
reader = tf.TFRecordReader()
files_queue = tf.train.string_input_producer(["./TFTest/test.tf"])

_, serialized_example = reader.read(files_queue) 

feature={
    'data': tf.FixedLenFeature([], tf.string), 
    'h': tf.FixedLenFeature([], tf.int64),
    'w': tf.FixedLenFeature([], tf.int64),
    'c': tf.FixedLenFeature([], tf.int64),
    'name': tf.FixedLenFeature([], tf.string),  
}
img_features = tf.parse_single_example(serialized_example,feature)
imagearr = tf.decode_raw(img_features["data"],tf.uint8)
imagearr = tf.reshape(imagearr, [299, 299, 3])
imageName = img_features["name"]

img_batch,name_batch = tf.train.batch([imagearr,imageName],batch_size=5, capacity=5)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    final_result_tensor = sess.graph.get_tensor_by_name("import/final_result:0")
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(sess=sess,coord=coord)
    plt.figure()
    for i in range(5):
        image_datas,image_names = sess.run([img_batch,name_batch])
        for j in range(len(image_datas)):
            image_data = image_datas[j]
            image_name = image_names[j].decode('utf-8')
            plt.title(image_name)
            plt.imshow(image_data)
            plt.axis("off")
            plt.show()
            image_data = image_data.astype(np.float32)
            image_data = (image_data/255)
            image_data = image_data[None,...]
            pre = sess.run([final_result_tensor],feed_dict={'import/Placeholder:0':image_data})
            pre = np.squeeze(pre)
            indexs = pre.argsort()[:-6:-1]
            for index in indexs:
                print("{}： probability is {:.2%}".format(id_to_classname(index),pre[index]))
    coord.request_stop()
    coord.join(threads)
```
