```python
import SimpleResNet
import tensorflow as tf
from tensorflow.contrib import slim
```

```python
datasets = "./datasets/flower_photos/"
HEIGHT = 244
WIDTH = 244
CHANNEL = 3
EPOCH = 10000
BATCH_SIZE = 32
```

```python
input_datas,label_datas,classmap = SimpleResNet.loadDataSet(datasets,WIDTH,HEIGHT,CHANNEL)
classNum = len(classmap)
```

```python
tf.reset_default_graph()

inputs = tf.placeholder(tf.float32,[None,WIDTH,HEIGHT,CHANNEL])
labels = tf.placeholder(tf.float32,[None,classNum])
is_training_batch_normal = tf.placeholder(tf.bool,name="isTrainBN")

with slim.arg_scope(SimpleResNet.resnet_arg_scope(is_training_batch_normal)):
    net, end_points,train_op = SimpleResNet.resnet_v2_50(inputs,labels,classNum)
```

```python
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    SimpleResNet.time_tensorflow_run(input_datas,label_datas,inputs,labels,sess, net,end_points,train_op, "Forward",EPOCH,BATCH_SIZE,None,100,1000)
```

```python
#上面训练到一半死机了，接着训练
init = tf.global_variables_initializer()
lastest_checkpoint = tf.train.latest_checkpoint("./ckpt")
with tf.Session() as sess:
    sess.run(init)
    SimpleResNet.time_tensorflow_run(input_datas,label_datas,inputs,labels,sess, net,end_points,train_op, "Forward",EPOCH,BATCH_SIZE,lastest_checkpoint,100,1000)
```

```python
#export pbtxt
with open("./output_labels.txt","w") as f:
    for index in range(classNum):
        f.write(classmap[index]+"\n")
```
