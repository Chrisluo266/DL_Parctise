```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
X,Y = make_blobs(1000,3,random_state=1000,centers=[[3,3, 3], [0,0,0], [1,1,1], [2,2,2]], cluster_std=[0.2, 0.1, 0.2, 0.2])
fig = plt.figure()
ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=30, azim=20)
plt.scatter(X[:, 0], X[:, 1], X[:, 2],marker='o',c=Y)
plt.show()
```

## PCA降维

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)   #大于1表示，保留的成分个数  3
pca.fit(X)
print("特征成分方差：",pca.explained_variance_)
print("特征成分方差比例：",pca.explained_variance_ratio_)
print("最终保留的特征数：",pca.n_components_)

pca2 = PCA(n_components=2)  #大于1表示，保留的成分个数   2
pca2.fit(X)
print("特征成分方差：",pca2.explained_variance_)
print("特征成分方差比例：",pca2.explained_variance_ratio_)
print("最终保留的特征数：",pca2.n_components_)

pca3 = PCA(n_components=0.99)  #小于1表示，先将特征按比例从高到低排序，然后从第一个开始逐渐相加，当和第一次大于这个参数值时停止，得到保留的特征。0.99>0.9825  0.99<0.9825 + 0.0092
pca3.fit(X)
print("特征成分方差：",pca3.explained_variance_)
print("特征成分方差比例：",pca3.explained_variance_ratio_)
print("最终保留的特征数：",pca3.n_components_)
```

```python
plt.figure()
X_new = pca2.transform(X)
plt.scatter(X_new[:,0],X_new[:,1],c=Y)
plt.show()
```
