```python
from sklearn import datasets
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

### 读取测试集

```python
loaded_data = datasets.load_boston()
data_x = loaded_data.data
data_y = loaded_data.target
print("data_x.shape = ",data_x.shape)
print("data_y.shape = ",data_y.shape)
```

### 模型

```python
model = LinearRegression()
train_x,test_x,train_y,test_y = train_test_split(data_x,data_y,test_size=0.1)
model.fit(train_x,train_y)
```

### 测试

```python
predict = model.predict(test_x)
print("predict =",predict)
print("real    =",test_y)
loss = test_y - predict
print("loss    =",loss)
total_loss = np.sum(np.abs(loss))
print("total_loss = ",total_loss)
```

### make datasets

```python
make_data_x,make_data_y = datasets.make_regression(100,1,10,1,noise=10)
print(make_data_x.shape)
print(make_data_y.shape)
model2 = LinearRegression()
model2.fit(make_data_x,make_data_y)
```

### plot regression

```python
import math
maxX = math.ceil(np.max(make_data_x))
minX = math.floor(np.min(make_data_x))
x_coords = np.arange(minX-1,maxX+1,(maxX-minX)/30)
plt.figure(figsize=(15,15))
plt.scatter(make_data_x,make_data_y,color='b')
y_coords = model2.predict(x_coords[...,None])
y_coords = np.squeeze(y_coords)
plt.plot(x_coords,y_coords,color='r')
plt.show()
```
