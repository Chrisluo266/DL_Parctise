```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
html = urlopen("https://www.bilibili.com/").read().decode("utf-8")
soup = BeautifulSoup(html,features="lxml")
```

```python
jpgs = soup.find_all("img",{"src":re.compile(".+\.jpg")})
for jpg in jpgs:
    print(jpg["src"])
```

```python
jpgs = soup.find_all("a",{"href":re.compile("^.+\/$")})
for jpg in jpgs:
    print(jpg["href"])
```
