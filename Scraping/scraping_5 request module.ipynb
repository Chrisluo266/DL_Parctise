# Post提交数据

# cookies

```python
data = {'username':'UUUU','password':'password'}
req = requests.post("http://pythonscraping.com/pages/cookies/welcome.php",data = data)
cookie_dict = req.cookies.get_dict()
print(cookie_dict)
print(req.text)
req = requests.get("http://pythonscraping.com/pages/cookies/profile.php",cookies = req.cookies)
print(req.text)
```

```python
params ={"username": "UUU","password": "password"}

session = requests.session()
session.cookies.update(cookie_dict)
post_obj = session.post("http://pythonscraping.com/pages/cookies/welcome.php", params)
s = session.get("http://pythonscraping.com/pages/cookies/profile.php")
print(post_obj.text)
print(s.text)
```
