```python
import requests
import os
from bs4 import BeautifulSoup
SAVE_DIR = "./download/test/"
os.makedirs(SAVE_DIR,exist_ok=True)
#URL = "http://www.nationalgeographic.com.cn/animals/"
BASE_URL = "http://www.ngchina.com.cn"
URL = BASE_URL + "/photography/photo_of_the_day/"

def download(url,path):
    if not os.path.exists(path):
        req = requests.get(url,stream=True)
        with open(path,"wb") as f:
            for chunk in req.iter_content(chunk_size=1024):
                f.write(chunk)
def htmlUrlToImageUrl(htmlUrl):
    req = requests.get(url)
    soup = BeautifulSoup(req.text,features="lxml")
    content = soup.find("div",{'class':"content js_content"})
    img = content.find("img")
    return img["src"]
```

```python
#查看更多
req = requests.get(URL)
html = req.text
soup = BeautifulSoup(html,features="lxml")
imglist_all = soup.find_all("ul",{"class":"cf showImg-list"})
for imglist in imglist_all:
    imga_all = imglist.find_all("a",{"class":'imgs'})
    for imga in imga_all:
        url = BASE_URL + imga['href']
        url = htmlUrlToImageUrl(url)
        imgName = url.split("/")[-1]
        download(url,os.path.join(SAVE_DIR,imgName))
        print(url)
```
